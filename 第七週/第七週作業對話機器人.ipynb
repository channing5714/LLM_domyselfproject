## 安裝必要套件

```python
import os
import gradio as gr
```

## 對話角色 system prompt

```python
customer_system_prompt = "你是一位對產品不滿的顧客，請用中文口語表達你的不滿，但不要罵人，並指出你希望的處理方式。"
agent_system_prompt = "你是一位有禮貌、冷靜且專業的客服人員，請安撫客戶並提出明確處理方法。請用清晰、有條理的中文回答。"
```

## 顧客模型呼叫

```python
def chat_with_groq_customer(user_input):
    try:
        response = openai.ChatCompletion.create(
            model="llama3-70b-8192",
            messages=[
                {"role": "system", "content": customer_system_prompt},
                {"role": "user", "content": user_input}
            ],
            temperature=0.8,
            top_p=0.95,
            max_tokens=256
        )
        return response['choices'][0]['message']['content'].strip()
    except Exception as e:
        return f"❌ 顧客模型出錯：{str(e)}"
```

## 客服模型呼叫

```python
def chat_with_groq_agent(user_input):
    response = openai.ChatCompletion.create(
        model="llama3-70b-8192",  # ✅ 客服用模型
        messages=[
            {"role": "system", "content": agent_system_prompt},
            {"role": "user", "content": user_input}
        ],
        temperature=0.7,
        top_p=0.9,
        max_tokens=512
    )
    return response['choices'][0]['message']['content'].strip()
```

## 自動執行三輪對話

```python
chat_history = []

def dual_chat(user_input):
    global chat_history
    customer_msg = user_input
    chat_history.append(f"🟢 顧客：{customer_msg}")

    for i in range(3):
        agent_msg = chat_with_groq_agent(customer_msg)
        chat_history.append(f"🔵 客服：{agent_msg}")

        customer_msg = chat_with_groq_customer(agent_msg)
        chat_history.append(f"🟢 顧客：{customer_msg}")

    return "\n\n".join(chat_history)

def clear_history():
    global chat_history
    chat_history = []
    return ""
```

## gradio 介面

```python
with gr.Blocks() as demo:
    gr.Markdown("## 🤖 雙 Groq 模型對話模擬器：顧客 & 客服")
    with gr.Row():
        user_input = gr.Textbox(label="顧客開場白", placeholder="例如：你們寄來的耳機有問題！")
        submit_btn = gr.Button("開始三輪對話")
        clear_btn = gr.Button("清除對話")
    chat_output = gr.Textbox(label="對話紀錄", lines=20)

    submit_btn.click(dual_chat, inputs=user_input, outputs=chat_output)
    clear_btn.click(clear_history, outputs=chat_output)

demo.launch()
```
